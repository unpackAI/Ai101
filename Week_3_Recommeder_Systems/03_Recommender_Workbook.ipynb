In order to access the files we need to download and unzip them, it can be done by utilizing `untar_data`. User ratings have been split between two spreadsheets, which is not convenient to train the model as it needs the entire dataset to give better predictions. We will have to first create a DataFrame for each file (`UserRatings1` and `UserRatings2`), and then concatenate them before we move to step 3.

We have provided some help for this data set, because of the way the data is set up.

```python
path = untar_data('https://www.dropbox.com/s/ef1mytjms1indlf/jester_collaborative_filtering.zip?dl=1')
Path.BASE_PATH = path
```

```python
path.ls()
```

```python
UserRatings1 = pd.read_csv(path/'UserRatings1.csv')
UserRatings1.head()
```

```python
UserRatings2 = pd.read_csv(path/'UserRatings2.csv')
UserRatings2.head()
```

```python
joke_text = pd.read_csv(path/'JokeText.csv')
joke_text.head()
```

```python
UserRatings2 = UserRatings2.drop('JokeId', axis=1)
UserRatings = pd.concat([UserRatings1, UserRatings2], axis=1)
UserRatings.iloc[:,36709:36720] 
```

#### 3. Carry out data transformation

```python
ratings = UserRatings.melt(id_vars="JokeId",var_name='users', value_name="rating")
```

```python
ratings.users = ratings.users.str.replace("User","")
```

```python
ratings.dropna(inplace=True)
```

```python
ratings = ratings.merge(joke_text, on="JokeId")
ratings.head()
```
